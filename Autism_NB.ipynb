```python
import time
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/pheno_train.csv")
print(data)
```

```python
data=data.fillna(1)
```

```python
data=pd.get_dummies(data)
```

```python
print(data)
```

```python
inputs = data.drop('SUB_IN_SMP',axis=1)
outputs = data['SUB_IN_SMP']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,test_size=0.25,random_state=0)
print("x_train\n",x_train)
print("x_test\n",x_test)
print("y_train\n",y_train)
print("y_test\n",y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
start_time = time.time()
nb=GaussianNB()
nb.fit(x_train,y_train)
end_time = time.time()
```

```python
print("Training time:",end_time-start_time)
```

```python
from sklearn.metrics import accuracy_score
y_pred=nb.predict(x_test)
y_pred
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.metrics import confusion_matrix,classification_report
a=confusion_matrix(y_test,y_pred)
b=classification_report(y_test,y_pred)
print(a)
print(b)
```
