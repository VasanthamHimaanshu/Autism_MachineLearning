```python
import time
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/pheno_train.csv')
print(data)
```

```python
data=data.fillna(0)
```

```python
data=pd.get_dummies(data)
```

```python
inputs = data.drop('SUB_IN_SMP', axis=1)
outputs = data['SUB_IN_SMP']
print(inputs)
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,outputs,test_size=0.2, random_state=42)
start_time=time.time()
regressor = RandomForestClassifier(n_estimators = 100, random_state = 0)
regressor.fit(inputs,outputs)
end_time=time.time()
```

```python
print("the time taken for the model is:",end_time-start_time)
```

```python
y_pred=regressor.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy:', accuracy)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
a=confusion_matrix(y_test,y_pred)
b=classification_report(y_test,y_pred)
print(a)
print(b)
```
