```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/autism.csv")
```

```python
data
```

```python
data = data.replace('?', -1)
```

```python
data
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
inputs=data.drop("Class_NO",axis=1)
outputs=data['Class_NO']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,test_size=0.25,random_state=0)
print("x_train\n",x_train)
print("x_test\n",x_test)
print("y_train\n",y_train)
print("y_test\n",y_test)
```

```python
from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.fit_transform(x_test)
print("x_train\n",x_train)
print("x_test\n",x_test)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
d=accuracy_score(y_test,y_pred)
print(d*100)
```
